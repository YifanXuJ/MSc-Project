# Review all label evaluation csv files to find the labeling method that yields the best results

```python
import numpy as np
import os
from matplotlib import pyplot as plt
import cv2
import csv
import pandas as pd
import glob
```

```python
from auto_label import PRECISION_50, PRECISION_X, FSCORE_SOLO, PRECISION_SOLO
```

# Use only the last 201 slice to label

```python
# Set the target segmentation results 
seg_model = ['gmm', 'k-means']   # choose between 'gmm' and 'k-means'
seg_nd = ['3d', '4d']  # choose between '3d' and '4d'
cluster_num = [16, 32, 64, 128]   # choose between 16, 32, 64, and 128
mode = [PRECISION_50, PRECISION_X, FSCORE_SOLO, PRECISION_SOLO]
threshold = [0.6, 0.7, 0.8, 0.9]

header = ['seg_model', 'seg_nd', 'cluster_num', 'mode', 'threshold', 'pore_precision', 'pore_recall', 'pore_f1', 'gypsum_precision', 'gypsum_recall', 'gypsum_f1', 
          'celestite_precision', 'celestite_recall', 'celestite_f1', 'bassanite_precision', 'bassanite_recall', 'bassanite_f1', 'macro_f1_4', 'macro_f1_3']

f = open('review_201_mode_all_train.csv', 'w')
writer = csv.writer(f)
writer.writerow(header)

for model in seg_model:
    for nd in seg_nd:
        for c in cluster_num:
            for m in mode:
                #corresponding label evaluation csv file
                if m == PRECISION_X:
                    for t in threshold:
                        csv_file = os.path.join(os.getcwd(), 'label_mode_evaluation_aligned', '{}_{}_{}_label_mode_{}_{}.csv'.format(model, nd, c, m, t))
                        df = pd.read_csv(csv_file)
                        # df = df.loc[df['slice'] < 500] # test set only
                        df = df.loc[df['slice'] >=  600]
                        stats = df.mean()
                        data = [model, nd, c, m, t] 
                        data.extend(stats[1:])
                        writer.writerow(data)
                else:
                    csv_file = os.path.join(os.getcwd(), 'label_mode_evaluation_aligned', '{}_{}_{}_label_mode_{}.csv'.format(model, nd, c, m))
                    df = pd.read_csv(csv_file)
                    # df = df.loc[df['slice'] < 500] # test set only
                    df = df.loc[df['slice'] >=  600]
                
                    if m == PRECISION_50:
                        data = [model, nd, c, m, 0.5] 

                    elif m == PRECISION_SOLO:
                        data = [model, nd, c, m, 0]
                    
                    else:
                        data = [model, nd, c, m, 1]
                    
                    stats = df.mean()
                    data.extend(stats[1:])
                    writer.writerow(data)

                    

f.close()
```
